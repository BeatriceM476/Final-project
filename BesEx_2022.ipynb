# Data cleaning

```python
# Liste des noms de colonnes
colonnes = [
    "Mediennummer",
    "Sigel",
    "Fachst_1",
    "Fachst_2",
    "Aufstellung",
    "Standort",
    "MTC",
    "Physikalische_Form",
    "Sachgr.",
    "Untergr.",
    "Umfang",
    "Beilagen",
    "Preis",
    "StO",
    "HH_Titel",
    "HH_Konto",
    "Inventarisierung",
    "Lieferant",
    "Ausleihen",
    "Verlängerungen",
    "Erwerbungsart",
    "Anzahl_Exemplare"
]

# Lecture du CSV avec séparateur "~" et noms de colonnes
BesEx_2022_df = pd.read_csv(
    "BESEX_2022.csv",
    sep="~",
    header=None,
    names=colonnes
)

BesEx_2022_df
```

```python
BesEx_2022_df.info()
```

```python
# Datatype str pour Mediennummer
BesEx_2022_df["Mediennummer"] = (
    BesEx_2022_df["Mediennummer"]
    .astype(float)  # au cas où il y ait encore du float
    .astype("Int64") # Pandas integer nullable
    .astype(str)    # en texte, pour éviter le .0
)
```

```python
# Datatype str pour HH-Titel
BesEx_2022_df["HH_Titel"] = (
    BesEx_2022_df["HH_Titel"]
    .astype(float)  # au cas où il y ait encore du float
    .astype("Int64") # Pandas integer nullable
    .astype(str)    # en texte, pour éviter le .0
)
```

```python
#effacer valeurs incorrectes dans H-Konto
BesEx_2022_df.loc[BesEx_2022_df['HH_Konto'] == "RABATT", 'HH_Konto'] = np.nan
BesEx_2022_df.loc[BesEx_2022_df['HH_Konto'] == "1 Poster Winkelgasse", 'HH_Konto'] = np.nan
```

```python
# Extraire uniquement les chiffres
BesEx_2022_df["HH_Konto"] = (
    BesEx_2022_df["HH_Konto"]
    .astype(str)                       # tout en texte pour traitement
    .str.extract(r'(\d+)')              # garde uniquement la partie numérique
    .astype("Int64")                    # entier nullable (Pandas)
)
BesEx_2022_df
```

```python
# conversion années en type datetime
BesEx_2022_df['Inventarisierung'] = (
    BesEx_2022_df['Inventarisierung']
    .astype('Int64')  # supprime le .0 tout en gérant les NaN
    .astype(str)      # en texte pour datetime
)

BesEx_2022_df['Inventarisierung'] = pd.to_datetime(
    BesEx_2022_df['Inventarisierung'],
    format='%Y%m%d',
    errors='coerce'
)
BesEx_2022_df
```

```python
# Nettoyage de la colonne Preis
BesEx_2022_df["Preis"] = (
    BesEx_2022_df["Preis"]
    .astype(str)                      # tout en texte pour manipuler
    .str.replace(",", ".", regex=False) # remplace virgule par point
    .str.replace(r"[^\d\.]", "", regex=True) # supprime tout sauf chiffres et point
    .replace("", np.nan)               # remplace les chaînes vides par NaN
    .astype(float)                     # convertit enfin en float
)
```

```python
# data types conversion
BesEx_2022_df.astype({
    'Sigel':'object',
    'Standort':'object',
    'MTC':'object',
    'Preis':'float',
    'Lieferant':'object'
}).dtypes
BesEx_2022_df = BesEx_2022_df.astype({
    'Sigel':'object',
    'Standort':'object',
    'MTC':'object',
    'Preis':'float',
    'Lieferant':'object'
})
BesEx_2022_df.dtypes
```

```python
# Liste des chaînes à traiter comme valeurs manquantes
fausses_nulls = ["<NA>", "NA", "N/A", "nan", "NaN", "None", "", "null", "NULL"]

# Sélectionner seulement les colonnes de type objet ou string
colonnes_objet = BesEx_2022_df.select_dtypes(include=["object", "string"]).columns

# Forcer ces colonnes à rester en string
BesEx_2022_df[colonnes_objet] = BesEx_2022_df[colonnes_objet].astype("string")

# Remplacer uniquement dans ces colonnes
BesEx_2022_df[colonnes_objet] = BesEx_2022_df[colonnes_objet].replace(fausses_nulls, pd.NA)

# Reforcer en string pour éviter toute reconversion
BesEx_2022_df[colonnes_objet] = BesEx_2022_df[colonnes_objet].astype("string")

print(BesEx_2022_df.dtypes)
```

```python
# delete null values
BesEx_2022_df.dropna(subset=['Mediennummer'], inplace=True)
BesEx_2022_df.dropna(subset=['Physikalische_Form'], inplace=True)
BesEx_2022_df.dropna(subset=['Sachgr.'], inplace=True)
print(BesEx_2022_df.isnull().sum())
```

```python
# delete colonnes
BesEx_2022_df = BesEx_2022_df.drop(columns=['Umfang', 'Beilagen', 'HH_Titel', 'HH_Konto', 'Lieferant', 'Standort', 'Untergr.'])
BesEx_2022_df.head(3)
```

```python
# Remplacer J par K dans Fachst_2
BesEx_2022_df['Fachst_2'] = BesEx_2022_df['Fachst_2'].replace('J', 'K')
```

```python
# test duplicate
duplicate_rows = BesEx_2022_df.duplicated().sum()
print(duplicate_rows)
# Pas de doublons.
```

```python
# comptabiliser les valeurs uniques
BesEx_2022_df.nunique()
```

```python
# créer une nouvelle colonne additionnant Ausleihen et Verlängerungen
BesEx_2022_df['Sum_Ausl'] = (
    BesEx_2022_df['Ausleihen'].fillna(0) +
    BesEx_2022_df['Verlängerungen'].fillna(0)
)
```
