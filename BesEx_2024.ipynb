# Data cleaning

```python
# delete colonnes
BesEx_2024_df = BesEx_2024_df.drop(columns=['Umfang', 'Beilagen', 'HH_Titel', 'HH_Konto', 'Lieferant', '???', 'Standort', 'Untergr.'])
BesEx_2024_df
```

```python
BesEx_2024_df.info()
```

```python
# Remplacer J par K dans Fachst_2
BesEx_2024_df['Fachst_2'] = BesEx_2024_df['Fachst_2'].replace('J', 'K')
```

```python
# comptabiliser les valeurs uniques
BesEx_2024_df.nunique()
```

```python
#test null values
print(BesEx_2024_df.isnull().sum())
```

```python
# delete null values
BesEx_2024_df.dropna(subset=['Mediennummer'], inplace=True)
BesEx_2024_df.dropna(subset=['Physikalische_Form'], inplace=True)
BesEx_2024_df.dropna(subset=['Sachgr.'], inplace=True)
print(BesEx_2024_df.isnull().sum())
```

```python
# test data types
BesEx_2024_df.dtypes
```

```python
# Datatype str pour Mediennummer
BesEx_2024_df["Mediennummer"] = (
    BesEx_2024_df["Mediennummer"]
    .astype(float)  # au cas où il y ait encore du float
    .astype("Int64") # Pandas integer nullable
    .astype(str)    # en texte, pour éviter le .0
)
```

```python
# conversion années en type datetime
BesEx_2024_df['Inventarisierung'] = (
    BesEx_2024_df['Inventarisierung']
    .astype('Int64')  # supprime le .0 tout en gérant les NaN
    .astype(str)      # en texte pour datetime
)

BesEx_2024_df['Inventarisierung'] = pd.to_datetime(
    BesEx_2024_df['Inventarisierung'],
    format='%Y%m%d',
    errors='coerce'
)
```

```python
# Nettoyage de la colonne Preis
BesEx_2024_df["Preis"] = (
    BesEx_2024_df["Preis"]
    .astype(str)                      # tout en texte pour manipuler
    .str.replace(",", ".", regex=False) # remplace virgule par point
    .str.replace(r"[^\d\.]", "", regex=True) # supprime tout sauf chiffres et point
    .replace("", np.nan)               # remplace les chaînes vides par NaN
    .astype(float)                     # convertit enfin en float
)
```

```python
# data types conversion
BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
}).dtypes
BesEx_2024_df = BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
})
BesEx_2024_df.dtypes
```

```python
# Liste des chaînes à traiter comme valeurs manquantes
fausses_nulls = ["<NA>", "NA", "N/A", "nan", "NaN", "None", "", "null", "NULL"]

# Sélectionner seulement les colonnes de type objet ou string
colonnes_objet = BesEx_2024_df.select_dtypes(include=["object", "string"]).columns

# Forcer ces colonnes à rester en string
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

# Remplacer uniquement dans ces colonnes
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].replace(fausses_nulls, pd.NA)

# Reforcer en string pour éviter toute reconversion
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

print(BesEx_2024_df.dtypes)
```

```python
# test duplicate
duplicate_rows = BesEx_2024_df.duplicated().sum()
print(duplicate_rows)
# Pas de doublons.
```

```python
# Konvertierung zu datetime (falls nicht schon gemacht)
BesEx_2024_df['Inventarisierung'] = pd.to_datetime(BesEx_2024_df['Inventarisierung'])

# Mapping E -> Erwachsene, K -> Kinder
mapping = {'E': 'Erwachsene', 'K': 'Kinder'}
BesEx_2024_df['Fachst_2'] = BesEx_2024_df['Fachst_2'].map(mapping)

# Neue Spalte Umsatz
BesEx_2024_df['Umsatz'] = BesEx_2024_df['Sum_Ausl'] / BesEx_2024_df['Anzahl_Exemplare']

# Gruppieren
umsatz_data = BesEx_2024_df.groupby(['Inventarisierung', 'Fachst_2'])['Umsatz'].mean().reset_index()

# Plot
plt.figure(figsize=(12,6))
sns.scatterplot(
    data=umsatz_data,
    x='Inventarisierung',
    y='Umsatz',
    hue='Fachst_2',
    marker='o'
)

plt.title("Umsatz nach Inventarisierung")
plt.xlabel("Inventarisierung")
plt.ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")

# Legendentitel ändern
plt.legend(title="Bestandsgruppe")

plt.tight_layout()
plt.show()
```

```python
# 1. S'assurer que la colonne Inventarisierung est en datetime
BesEx_2024_df['Inventarisierung'] = pd.to_datetime(BesEx_2024_df['Inventarisierung'], format="%Y-%m")

# 2. Créer une colonne Année-Mois
BesEx_2024_df['JahrMonat'] = BesEx_2024_df['Inventarisierung'].dt.to_period('M').dt.to_timestamp()

# 3. Agréger par mois et groupe Fachst_2
ausl_monatlich = (
    BesEx_2024_df
    .groupby(['JahrMonat', 'Fachst_2'])['Sum_Ausl']
    .sum()
    .reset_index()
)

# 4. Calculer le lissage 12 mois
ausl_monatlich['Sum_Ausl_roll'] = (
    ausl_monatlich
    .groupby('Fachst_2')['Sum_Ausl']
    .transform(lambda x: x.rolling(12, min_periods=1).mean())
)

# 5. Définir les dates limites
start_date = "2000-01-01"
end_date = "2023-12-31"
ausl_monatlich_filtered = ausl_monatlich[
    (ausl_monatlich['JahrMonat'] >= pd.to_datetime(start_date)) &
    (ausl_monatlich['JahrMonat'] <= pd.to_datetime(end_date))
]

# 6. Plot
plt.figure(figsize=(14,6))
sns.lineplot(
    data=ausl_monatlich_filtered,
    x="JahrMonat",
    y="Sum_Ausl_roll",
    hue="Fachst_2"
)
plt.title("Ausleihe nach Inventarisierungsdatum (12-Monats-Durchschnitt)")
plt.xlabel("Jahr")
plt.ylabel("Ausleihe")
plt.legend(title="Bestandsgruppe")
plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))
plt.tight_layout()
plt.show()
```

```python
# 1. S'assurer que la colonne date est en datetime
BesEx_2024_df['Inventarisierung'] = pd.to_datetime(BesEx_2024_df['Inventarisierung'], format="%Y-%m")

# 2. Créer une colonne Année-Mois
BesEx_2024_df['JahrMonat'] = BesEx_2024_df['Inventarisierung'].dt.to_period('M').dt.to_timestamp()

# Agréger par mois et groupe, moyenne de Umsatz
umsatz_monatlich = (
    BesEx_2024_df
    .groupby(['JahrMonat', 'Fachst_2'])['Umsatz']
    .mean()  # <- ici, on prend la moyenne
    .reset_index()
)

# Moyenne mobile 12 mois pour lisser
umsatz_monatlich['Umsatz_roll'] = (
    umsatz_monatlich
    .groupby('Fachst_2')['Umsatz']
    .transform(lambda x: x.rolling(12, min_periods=1).mean())
)

# 5. Définir les dates limites
start_date = "2000-01-01"
end_date = "2023-12-31"

umsatz_monatlich_filtered = umsatz_monatlich[
    (umsatz_monatlich["JahrMonat"] >= start_date) &
    (umsatz_monatlich["JahrMonat"] <= end_date)
]

# 6. Tracer le plot
plt.figure(figsize=(14,6))
sns.lineplot(
    data=umsatz_monatlich_filtered,
    x="JahrMonat",
    y="Umsatz_roll",
    hue="Fachst_2"
)
plt.title("Umsatz nach Inventarisierungsdatum")
plt.xlabel("Jahr")
plt.ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")
plt.legend(title="Bestandsgruppe")
plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))
plt.tight_layout()
plt.show()
```

Umsatz nach Erwerbungsart

```python
# ---------------------------
# 1️⃣ Préparation des données pour le graphique 1 (Erwerbungsart)
# ---------------------------
filtered_df = BesEx_2024_df[BesEx_2024_df['Erwerbungsart'].isin(['E', 'G', 'K'])]
filtered_df['Umsatz'] = filtered_df['Sum_Ausl'] / filtered_df['Anzahl_Exemplare']

erwerbung_data = filtered_df.groupby('Erwerbungsart')['Umsatz'].mean().reset_index()

label_mapping_erwerbung = {'E': 'Ersatz', 'G': 'Geschenk', 'K': 'Kauf'}
erwerbung_data['Erwerbungsart_Label'] = erwerbung_data['Erwerbungsart'].map(label_mapping_erwerbung)

# ---------------------------
# 2️⃣ Préparation des données pour le graphique 2 (StO)
# ---------------------------
BesEx_2024_df['Umsatz'] = BesEx_2024_df['Sum_Ausl'] / BesEx_2024_df['Anzahl_Exemplare']
BesEx_2024_df['StO_clean'] = BesEx_2024_df['StO'].fillna('NaN')

sto_data = BesEx_2024_df.groupby('StO_clean')['Umsatz'].mean().reset_index()

label_mapping_sto = {'NaN': 'Direktkauf', 'S': 'Standing Order'}
sto_data['StO_Label'] = sto_data['StO_clean'].map(label_mapping_sto)

# ---------------------------
# 3️⃣ Visualisation en subplots
# ---------------------------
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graphique 1 : Erwerbungsart
sns.barplot(
    data=erwerbung_data,
    x='Erwerbungsart_Label',
    y='Umsatz',
    palette='tab10',
    ax=axes[0]
)
axes[0].set_title("Umsatz nach Erwerbungsart")
axes[0].set_xlabel("Erwerbungsart")
axes[0].set_ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")

# Graphique 2 : StO
sns.barplot(
    data=sto_data,
    x='StO_Label',
    y='Umsatz',
    palette='tab10',
    ax=axes[1]
)
axes[1].set_title("Umsatz nach Kaufart")
axes[1].set_xlabel("Kaufart")
axes[1].set_ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")

plt.tight_layout()
plt.show()
```

Effizienz = Ausleihanteil / Bestandsanteil
Effizienz = 1 bedeutet: Die Bestandsgruppe entspricht dem Durchschnitt dieser Bibliothek
• Effizienz < 1 bedeutet: Die Bestandsgruppe wird unterdurchschnittlich genutzt = Bestand verkleinern!
• Effizienz > 1 bedeutet: Die Bestandsgruppe wird überdurchschnittlich genutzt = Bestand vergrößern!

```python
# 1️⃣ Filtrer MTC=1 et Erwachsene
df_filtered = BesEx_2024_df[
    (BesEx_2024_df['MTC'].astype(str).str.strip() == "1") &
    (BesEx_2024_df['Fachst_2'].astype(str).str.strip() == "Erwachsene")
]

# 2️⃣ Taux moyen global
global_mean = df_filtered['Umsatz'].mean()

# 3️⃣ Effizienz par Sachgruppe
effizienz_df = df_filtered.groupby('Sachgr.').agg({
    'Umsatz': 'mean',              # moyenne par Sachgruppe
    'Mediennummer': 'count'        # nombre d'exemplaires
}).rename(columns={'Mediennummer': 'Anzahl_Exemplare'}).reset_index()

# 4️⃣ Calculer Effizienz
effizienz_df['Effizienz'] = (effizienz_df['Umsatz'] / global_mean).round(2)

# 5️⃣ Garder seulement certaines Sachgruppen
sachgruppen_keep = [
    'Gesch', 'Nat', 'Roman', 'Ku', 'Lit', 'Phil', 'R', 'Rel', 'Allg',
    'Dramen', 'Werke', 'Vo', 'Gedichte', 'Märchen', 'Psy', 'Mus', 'B',
    'Lan', 'Pä', 'Med', 'Pol', 'Soz', 'Spra', 'Bio', 'HW', 'Erzählungen',
    'Th', 'Comic', 'Bran', 'Spo', 'Essays', 'SoP', 'Erd', 'Sagen', 'Ma',
    'BWL', 'Tech', 'Wi', 'BID', 'Sammlungen', 'Cartoon', 'EDV',
    'Kabarett', 'Aphorismen', 'Witze'
]
effizienz_df = effizienz_df[effizienz_df['Sachgr.'].isin(sachgruppen_keep)]

# 6️⃣ Plot
plt.figure(figsize=(18,6))

scatter = plt.scatter(
    x=effizienz_df['Umsatz'],
    y=effizienz_df['Effizienz'],
    s=effizienz_df['Anzahl_Exemplare']*0.5,  # Blasengröße
    c=effizienz_df['Effizienz'],
    cmap='viridis',
    alpha=0.7
)

# Linie Effizienz = 1
plt.axhline(1, color='grey', linestyle='--', linewidth=1)

# Labels
texts = []
for i, (idx, row) in enumerate(effizienz_df.iterrows()):
    offset = 0.02 * (-1)**i
    texts.append(
        plt.text(
            row['Umsatz'],
            row['Effizienz'] + offset,
            row['Sachgr.'],
            ha='center',
            va='bottom' if offset > 0 else 'top',
            fontsize=8,
            color='black'
        )
    )
adjust_text(texts, arrowprops=dict(arrowstyle='-', color='grey', alpha=0.5))

# Legende für Blasengröße
for size in [50, 200, 500]:
    plt.scatter([], [], c='grey', alpha=0.6, s=size, label=f'{int(size/0.5)} Exemplare')
plt.legend(title='Anzahl Exemplare', scatterpoints=1, frameon=True, fontsize=9)

# Farbskala
cbar = plt.colorbar(scatter)
cbar.set_label("Effizienz")

# Achsen & Titel
plt.xlabel("Umsatz")
plt.ylabel("Effizienz")
plt.title("Effizienz nach Sachgruppe (MTC=1, Erwachsene)")

plt.tight_layout()
plt.show()
```
