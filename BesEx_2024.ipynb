# Librairies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data wrangling

```python
# Gathering data
BesEx_2024_df = pd.read_csv('BESEX_2024.csv', sep='~')
BesEx_2024_df
```

```python
# Liste des noms de colonnes
colonnes = [
    "Mediennummer",
    "Sigel",
    "Fachst_1",
    "Fachst_2",
    "Aufstellung",
    "Standort",
    "MTC",
    "Physikalische_Form",
    "Sachgr.",
    "Untergr.",
    "Umfang",
    "Beilagen",
    "Preis",
    "StO",
    "HH_Titel",
    "HH_Konto",
    "Inventarisierung",
    "Lieferant",
    "???",
    "Ausleihen",
    "Verlängerungen",
    "Erwerbungsart",
    "Anzahl_Exemplare"
]

# Lecture du CSV avec séparateur "~" et noms de colonnes
BesEx_2024_df = pd.read_csv(
    "BESEX_2024.csv",
    sep="~",
    header=None,
    names=colonnes
)

BesEx_2024_df
```

```python
BesEx_2024_df.shape
```

# Data cleaning

```python
# delete colonnes
BesEx_2024_df = BesEx_2024_df.drop(columns=['Umfang', 'Beilagen', 'HH_Titel', 'HH_Konto', 'Lieferant', '???', 'Standort', 'Untergr.'])
BesEx_2024_df
```

```python
BesEx_2024_df.info()
```

```python
# Remplacer J par K dans Fachst_2
BesEx_2024_df['Fachst_2'] = BesEx_2024_df['Fachst_2'].replace('J', 'K')
```

```python
# comptabiliser les valeurs uniques
BesEx_2024_df.nunique()
```

```python
#test null values
print(BesEx_2024_df.isnull().sum())
```

```python
# delete null values
BesEx_2024_df.dropna(subset=['Mediennummer'], inplace=True)
BesEx_2024_df.dropna(subset=['Physikalische_Form'], inplace=True)
BesEx_2024_df.dropna(subset=['Sachgr.'], inplace=True)
print(BesEx_2024_df.isnull().sum())
```

```python
# test data types
BesEx_2024_df.dtypes
```

```python
# Datatype str pour Mediennummer
BesEx_2024_df["Mediennummer"] = (
    BesEx_2024_df["Mediennummer"]
    .astype(float)  # au cas où il y ait encore du float
    .astype("Int64") # Pandas integer nullable
    .astype(str)    # en texte, pour éviter le .0
)
```

```python
# conversion années en type datetime
BesEx_2024_df['Inventarisierung'] = (
    BesEx_2024_df['Inventarisierung']
    .astype('Int64')  # supprime le .0 tout en gérant les NaN
    .astype(str)      # en texte pour datetime
)

BesEx_2024_df['Inventarisierung'] = pd.to_datetime(
    BesEx_2024_df['Inventarisierung'],
    format='%Y%m%d',
    errors='coerce'
)
```

```python
# Nettoyage de la colonne Preis
BesEx_2024_df["Preis"] = (
    BesEx_2024_df["Preis"]
    .astype(str)                      # tout en texte pour manipuler
    .str.replace(",", ".", regex=False) # remplace virgule par point
    .str.replace(r"[^\d\.]", "", regex=True) # supprime tout sauf chiffres et point
    .replace("", np.nan)               # remplace les chaînes vides par NaN
    .astype(float)                     # convertit enfin en float
)
```

```python
# data types conversion
BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
}).dtypes
BesEx_2024_df = BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
})
BesEx_2024_df.dtypes
```

```python
# Liste des chaînes à traiter comme valeurs manquantes
fausses_nulls = ["<NA>", "NA", "N/A", "nan", "NaN", "None", "", "null", "NULL"]

# Sélectionner seulement les colonnes de type objet ou string
colonnes_objet = BesEx_2024_df.select_dtypes(include=["object", "string"]).columns

# Forcer ces colonnes à rester en string
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

# Remplacer uniquement dans ces colonnes
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].replace(fausses_nulls, pd.NA)

# Reforcer en string pour éviter toute reconversion
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

print(BesEx_2024_df.dtypes)
```

```python
# test duplicate
duplicate_rows = BesEx_2024_df.duplicated().sum()
print(duplicate_rows)
# Pas de doublons.
```

```python
# comptabiliser les valeurs uniques
BesEx_2024_df.nunique()
```

```python
# créer une nouvelle colonne additionnant Ausleihen et Verlängerungen
BesEx_2024_df['Sum_Ausl'] = (
    BesEx_2024_df['Ausleihen'].fillna(0) +
    BesEx_2024_df['Verlängerungen'].fillna(0)
)
BesEx_2024_df
```

# Exploratory Data Analysis (EDA) 

Pas de corrélation

Ausleihzahl nach Inventarisierungsdatum

```python
# ---------------------------
# 1️⃣ S'assurer que Inventarisierung est en datetime
# ---------------------------
BesEx_2024_df['Inventarisierung'] = pd.to_datetime(BesEx_2024_df['Inventarisierung'])

# ---------------------------
# 1️⃣bis Filtrer les données à partir de 2004-01-01
# ---------------------------
BesEx_2024_df = BesEx_2024_df[BesEx_2024_df['Inventarisierung'] >= '2004-01-01']

# ---------------------------
# 2️⃣ Calculer l'Umsatz et grouper par Inventarisierung et Fachst_2
# ---------------------------
# Calculer l'Umsatz (taux de rotation)
BesEx_2024_df['Umsatz'] = BesEx_2024_df['Sum_Ausl'] / BesEx_2024_df['Anzahl_Exemplare']

# Grouper par Inventarisierung et Fachst_2, puis calculer la moyenne de l'Umsatz
umsatz_data = BesEx_2024_df.groupby(['Inventarisierung', 'Fachst_2'])['Umsatz'].mean().reset_index()

# ---------------------------
# 3️⃣ Visualisation
# ---------------------------
plt.figure(figsize=(12,6))
sns.scatterplot(
    data=umsatz_data,
    x='Inventarisierung',
    y='Umsatz',
    hue='Fachst_2',
    marker='o'
)
plt.title("Umsatz nach Inventarisierung")
plt.xlabel("Inventarisierung")
plt.ylabel("Umsatz (Sum_Ausl / Anzahl_Exemplare)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------
# 1️⃣ Grouper par Erwerbungsart et sommer Sum_Ausl
# ---------------------------
erwerbung_data = BesEx_2024_df.groupby('Erwerbungsart')['Sum_Ausl'].sum().reset_index()

# ---------------------------
# 2️⃣ Visualisation
# ---------------------------
plt.figure(figsize=(12,6))
sns.barplot(
    data=erwerbung_data,
    x='Erwerbungsart',
    y='Sum_Ausl',
    #palette='viridis'  # Tu peux changer la palette de couleurs si tu veux
)
plt.title("Somme des emprunts (Sum_Ausl) par type d'acquisition (Erwerbungsart)")
plt.xlabel("Erwerbungsart")
plt.ylabel("Sum_Ausl")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------
# 1️⃣ Filtrer pour E, G, et K et calculer l'Umsatz
# ---------------------------
filtered_df = BesEx_2024_df[BesEx_2024_df['Erwerbungsart'].isin(['E', 'G', 'K'])]
filtered_df['Umsatz'] = filtered_df['Sum_Ausl'] / filtered_df['Anzahl_Exemplare']

# ---------------------------
# 2️⃣ Grouper par Erwerbungsart et calculer la moyenne de l'Umsatz
# ---------------------------
erwerbung_data = filtered_df.groupby('Erwerbungsart')['Umsatz'].mean().reset_index()

# ---------------------------
# 3️⃣ Visualisation
# ---------------------------
plt.figure(figsize=(10,6))
sns.barplot(
    data=erwerbung_data,
    x='Erwerbungsart',
    y='Umsatz',
    palette='viridis'  # Tu peux changer la palette de couleurs si tu veux
)
plt.title("Umsatz nach Erwerbungsart (E, G, K)")
plt.xlabel("Erwerbungsart")
plt.ylabel("Umsatz (Sum_Ausl / Anzahl_Exemplare)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------
# 1️⃣ Calculer l'Umsatz et vérifier les valeurs StO
# ---------------------------
BesEx_2024_df['Umsatz'] = BesEx_2024_df['Sum_Ausl'] / BesEx_2024_df['Anzahl_Exemplare']

# Remplacer les NaN par 'NaN' pour les rendre visibles
BesEx_2024_df['StO_clean'] = BesEx_2024_df['StO'].fillna('NaN')

# ---------------------------
# 2️⃣ Grouper par StO_clean et calculer la moyenne de l'Umsatz
# ---------------------------
sto_data = BesEx_2024_df.groupby('StO_clean')['Umsatz'].mean().reset_index()

# ---------------------------
# 3️⃣ Visualisation
# ---------------------------
plt.figure(figsize=(8,6))
sns.barplot(
    data=sto_data,
    x='StO_clean',
    y='Umsatz',
    palette='viridis'
)
plt.title("Umsatz nach Standing Order")
plt.xlabel("StO")
plt.ylabel("Umsatz (Sum_Ausl / Anzahl_Exemplare)")
plt.tight_layout()
plt.show()
```
