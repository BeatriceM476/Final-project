# Data wrangling

```python
# Gathering data
BesEx_2024_df = pd.read_csv('BESEX_2024.csv', sep='~')
BesEx_2024_df
```

```python
# Liste des noms de colonnes
colonnes = [
    "Mediennummer",
    "Sigel",
    "Fachst_1",
    "Fachst_2",
    "Aufstellung",
    "Standort",
    "MTC",
    "Physikalische_Form",
    "Sachgr.",
    "Untergr.",
    "Umfang",
    "Beilagen",
    "Preis",
    "StO",
    "HH_Titel",
    "HH_Konto",
    "Inventarisierung",
    "Lieferant",
    "???",
    "Ausleihen",
    "Verlängerungen",
    "Erwerbungsart",
    "Anzahl_Exemplare"
]

# Lecture du CSV avec séparateur "~" et noms de colonnes
BesEx_2024_df = pd.read_csv(
    "BESEX_2024.csv",
    sep="~",
    header=None,
    names=colonnes
)

BesEx_2024_df
```

# Data cleaning

```python
# delete colonnes
BesEx_2024_df = BesEx_2024_df.drop(columns=['Umfang', 'Beilagen', 'HH_Titel', 'HH_Konto', 'Lieferant', '???', 'Standort', 'Untergr.'])
BesEx_2024_df
```

```python
BesEx_2024_df.info()
```

```python
# Remplacer J par K dans Fachst_2
BesEx_2024_df['Fachst_2'] = BesEx_2024_df['Fachst_2'].replace('J', 'K')
```

```python
# comptabiliser les valeurs uniques
BesEx_2024_df.nunique()
```

```python
#test null values
print(BesEx_2024_df.isnull().sum())
```

```python
# delete null values
BesEx_2024_df.dropna(subset=['Mediennummer'], inplace=True)
BesEx_2024_df.dropna(subset=['Physikalische_Form'], inplace=True)
BesEx_2024_df.dropna(subset=['Sachgr.'], inplace=True)
print(BesEx_2024_df.isnull().sum())
```

```python
# test data types
BesEx_2024_df.dtypes
```

```python
# Datatype str pour Mediennummer
BesEx_2024_df["Mediennummer"] = (
    BesEx_2024_df["Mediennummer"]
    .astype(float)  # au cas où il y ait encore du float
    .astype("Int64") # Pandas integer nullable
    .astype(str)    # en texte, pour éviter le .0
)
```

```python
# conversion années en type datetime
BesEx_2024_df['Inventarisierung'] = (
    BesEx_2024_df['Inventarisierung']
    .astype('Int64')  # supprime le .0 tout en gérant les NaN
    .astype(str)      # en texte pour datetime
)

BesEx_2024_df['Inventarisierung'] = pd.to_datetime(
    BesEx_2024_df['Inventarisierung'],
    format='%Y%m%d',
    errors='coerce'
)
```

```python
# Nettoyage de la colonne Preis
BesEx_2024_df["Preis"] = (
    BesEx_2024_df["Preis"]
    .astype(str)                      # tout en texte pour manipuler
    .str.replace(",", ".", regex=False) # remplace virgule par point
    .str.replace(r"[^\d\.]", "", regex=True) # supprime tout sauf chiffres et point
    .replace("", np.nan)               # remplace les chaînes vides par NaN
    .astype(float)                     # convertit enfin en float
)
```

```python
# data types conversion
BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
}).dtypes
BesEx_2024_df = BesEx_2024_df.astype({
    'Sigel':'object',
    'MTC':'object',
    'Preis':'float'
})
BesEx_2024_df.dtypes
```

```python
# Liste des chaînes à traiter comme valeurs manquantes
fausses_nulls = ["<NA>", "NA", "N/A", "nan", "NaN", "None", "", "null", "NULL"]

# Sélectionner seulement les colonnes de type objet ou string
colonnes_objet = BesEx_2024_df.select_dtypes(include=["object", "string"]).columns

# Forcer ces colonnes à rester en string
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

# Remplacer uniquement dans ces colonnes
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].replace(fausses_nulls, pd.NA)

# Reforcer en string pour éviter toute reconversion
BesEx_2024_df[colonnes_objet] = BesEx_2024_df[colonnes_objet].astype("string")

print(BesEx_2024_df.dtypes)
```

```python
# test duplicate
duplicate_rows = BesEx_2024_df.duplicated().sum()
print(duplicate_rows)
# Pas de doublons.
```

```python
# comptabiliser les valeurs uniques
BesEx_2024_df.nunique()
```

```python
# créer une nouvelle colonne additionnant Ausleihen et Verlängerungen
BesEx_2024_df['Sum_Ausl'] = (
    BesEx_2024_df['Ausleihen'].fillna(0) +
    BesEx_2024_df['Verlängerungen'].fillna(0)
)
BesEx_2024_df
```

```python
# 1. S'assurer que la colonne date est en datetime
BesEx_2024_df['Inventarisierung'] = pd.to_datetime(BesEx_2024_df['Inventarisierung'], format="%Y-%m")

# 2. Créer une colonne Année-Mois
BesEx_2024_df['JahrMonat'] = BesEx_2024_df['Inventarisierung'].dt.to_period('M').dt.to_timestamp()

# Agréger par mois et groupe, moyenne de Umsatz
umsatz_monatlich = (
    BesEx_2024_df
    .groupby(['JahrMonat', 'Fachst_2'])['Umsatz']
    .mean()  # <- ici, on prend la moyenne
    .reset_index()
)

# Moyenne mobile 12 mois pour lisser
umsatz_monatlich['Umsatz_roll'] = (
    umsatz_monatlich
    .groupby('Fachst_2')['Umsatz']
    .transform(lambda x: x.rolling(12, min_periods=1).mean())
)

# 5. Définir les dates limites
start_date = "2000-01-01"
end_date = "2023-12-31"

umsatz_monatlich_filtered = umsatz_monatlich[
    (umsatz_monatlich["JahrMonat"] >= start_date) &
    (umsatz_monatlich["JahrMonat"] <= end_date)
]

# 6. Tracer le plot
plt.figure(figsize=(14,6))
sns.lineplot(
    data=umsatz_monatlich_filtered,
    x="JahrMonat",
    y="Umsatz_roll",
    hue="Fachst_2"
)
plt.title("Umsatz nach Inventarisierungsdatum")
plt.xlabel("Jahr")
plt.ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")
plt.legend(title="Bestandsgruppe")
plt.xlim(pd.to_datetime(start_date), pd.to_datetime(end_date))
plt.tight_layout()
plt.show()
```

Umsatz nach Erwerbungsart

```python
# ---------------------------
# 1️⃣ Préparation des données pour le graphique 1 (Erwerbungsart)
# ---------------------------
filtered_df = BesEx_2024_df[BesEx_2024_df['Erwerbungsart'].isin(['E', 'G', 'K'])]
filtered_df['Umsatz'] = filtered_df['Sum_Ausl'] / filtered_df['Anzahl_Exemplare']

erwerbung_data = filtered_df.groupby('Erwerbungsart')['Umsatz'].mean().reset_index()

label_mapping_erwerbung = {'E': 'Ersatz', 'G': 'Geschenk', 'K': 'Kauf'}
erwerbung_data['Erwerbungsart_Label'] = erwerbung_data['Erwerbungsart'].map(label_mapping_erwerbung)

# ---------------------------
# 2️⃣ Préparation des données pour le graphique 2 (StO)
# ---------------------------
BesEx_2024_df['Umsatz'] = BesEx_2024_df['Sum_Ausl'] / BesEx_2024_df['Anzahl_Exemplare']
BesEx_2024_df['StO_clean'] = BesEx_2024_df['StO'].fillna('NaN')

sto_data = BesEx_2024_df.groupby('StO_clean')['Umsatz'].mean().reset_index()

label_mapping_sto = {'NaN': 'Direktkauf', 'S': 'Standing Order'}
sto_data['StO_Label'] = sto_data['StO_clean'].map(label_mapping_sto)

# ---------------------------
# 3️⃣ Visualisation en subplots
# ---------------------------
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graphique 1 : Erwerbungsart
sns.barplot(
    data=erwerbung_data,
    x='Erwerbungsart_Label',
    y='Umsatz',
    palette='viridis',
    ax=axes[0]
)
axes[0].set_title("Umsatz nach Erwerbungsart")
axes[0].set_xlabel("Erwerbungsart")
axes[0].set_ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")

# Graphique 2 : StO
sns.barplot(
    data=sto_data,
    x='StO_Label',
    y='Umsatz',
    palette='viridis',
    ax=axes[1]
)
axes[1].set_title("Umsatz nach Kaufart")
axes[1].set_xlabel("Kaufart")
axes[1].set_ylabel("Umsatz (Ausleihzahl / Exemplarzahl)")

plt.tight_layout()
plt.show()
```
